```python
# Generate timestamps
import datetime
# Contains hashing algorithms
import hashlib
```

```python
# Defining the 'block' data sctructures
class Block:
    #each block has 7 attributes
    #1 number of the block
    #2 what data is stored in this block?
    #3 pointer to the next block
    #4 The hash of this block (serves as a unique ID and verifies...
    #5 nonce is a number only used once
    #6 store the hash (ID) of the previous block in the chain
    #7 timestamp
    blockNo = 0
    data = None
    next = None
    hash = None
    nonce = 0 # Number only used once
    previous_hash = 0x0
    timestamp = datetime.datetime.now()
    
    # Initializing a block by storing some data in it
    def __init__(self,data):
        self.data = data
    
    # Function to compute 'hash' of a block
    def hash(self):
        h = hashlib.sha256() # generates a 256bit signature
        h.update(
            str(self.nonce).encode('utf-8') +
            str(self.data).encode('utf-8') +
            str(self.previous_hash).encode('utf-8') +
            str(self.timestamp).encode('utf-8') +
            str(self.blockNo).encode('utf-8') 
        )
        return h.hexdigest()
    
    def __str__(self):
        # Print out the value of a block
        return "Block Hash: " + str(self.hash()) + "\nBlockNo: " + str(self.blockNo)
    
    
    
    
     
```

```python
# Create a Block Chain

# Define the blockchain datastructure
#consist of 'blocks' linked together
class Blockchain:
    
    maxNonce = 2**2
    diff = 10 # to compute our target hash
    target = 2 ** (256-diff)
    
    # Generate the first block in the block chain
    block = Block("Genesis")
    # Set it as the head of our block chain
    head = block
    
    #adds a given block to the chain of blocks
    def add(self, block):
        block.previous_hash = self.block.hash()
        block.blockNo = self.block.blockNo + 1
        
        self.block.next = block
        self.block = self.block.next
        
    # Determine whether or not we can add a given block to the blockchain
    def mine(self, block):
        for n in range(self.maxNonce):
            if int(block.hash(), 16) <= self.target:
                self.add(block)
                print(block)
                break
            else:
                block.nonce +=1
            
        
        
        
    
    
```
